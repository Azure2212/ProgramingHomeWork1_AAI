<h1>Basic PyTorch Question - 10 points</h1>
<h1>a. Construct an n-dimensional tensor in PyTorch.</h1>

```python
def nd_tensor_construct(listValue, shape, min=0, max=100):
    if len(listValue) != 0:
        tensor = torch.tensor(listValue)
        tensor = torch.reshape(tensor, shape)
    else:  # Random create
        tensor = torch.rand(size=shape) * (max - min) + min
    return tensor
```

<h1>b. Compute the sum, mean, or standard deviation of all elements in a tensor.<h1>

```python
def compute_sum_mean_std(tensor, typeCal = 3): # 0 is only calculate sum, 1 is only calculate mean_val, 2 is only calculate std_val, 3 is calculate all
    total_sum = tensor.sum()
    mean_val = tensor.float().mean()
    std_val  = tensor.float().std()

    if typeCal == 0: return total_sum
    elif typeCal == 1: return mean_val
    elif typeCal == 2: return std_val
    else: return total_sum, mean_val, std_val
```

<h1>c. Perform mathematical operations on tensors, such as finding the maximum or minimum value, or taking the dot product of two tensors.<h1>

```python
def maximum_in_tensor(tensor):
  return tensor.max()

def minimum_in_tensor(tensor):
  return tensor.min()

def dot_2_tensors(tensor1, tensor2):
  if tensor1.shape != tensor2.shape:
    raise ValueError("Shapes must match")
  return torch.sum(tensor1 * tensor2)
```

<h1>d.Access and manipulate specific elements or slices of a tensor, and index and slice tensors based on certain conditions.</h1>

```python
def access_element(tensor, row_idx, col_idx):
  print(tensor[row_idx, col_idx].numpy())

def manipulate_element(tensor, row_idx, col_idx, newValue):
  tensor[row_idx, col_idx] = newValue
  print(f"Assigned newValue = {newValue} for element at [{row_idx},{col_idx}] : {tensor[row_idx, col_idx]}")
  return tensor

def access_by_slice(tensor, row_idx_start, row_idx_end, col_idx_start, col_idx_end):
  print(tensor[row_idx_start:row_idx_end, col_idx_start: col_idx_end])
```

<h1>e. Define and call functions in PyTorch, and pass tensors as arguments to functions.</h1>

```python
def sum_2tensors(a, b):
  if a.shape != b.shape:
    raise Exception("tensor A's shape must be equal to tensor B'shape")
  return a + b

def minus_2tensors(a, b):
  if a.shape != b.shape:
    raise Exception("tensor A's shape must be equal to tensor B'shape")
  return a - b

def mul_2tensors(a, b):
  if a.shape != b.shape:
    raise Exception("tensor A's shape must be equal to tensor B'shape")
  return a * b
```

<h1>f. Perform matrix multiplication of m × n matrix with n × 1 vector to calculate m × 1 vector.</h1>

```python
def MatrixMul(a, b):
  if a.shape[1] != b.shape[0]:
    raise Exception("Col of tensor A should be equal to row of tensor B")
  return a @ b
```

<h1>g. Write a Python program to perform Hadamard product of m × k matrices.</h1>

```python
def hadamard_product(A, B):
    if A.shape != B.shape:
        raise ValueError("The shape didn't fit")
    return A * B # torch.mul(A,B)
```
