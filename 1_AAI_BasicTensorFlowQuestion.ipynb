<h1>Part2 Basic TensorFlow Question 10 points<h1>
<h1>a. Construct an n-dimensional tensor in TensorFlow.<h1>


```python
def nd_tensor_construct(listValue, shape, min = 0, max=100):
    if len(listValue) != 0: #For input from keyboard
      tensor = tf.constant(listValue)
      tensor = tf.reshape(tensor, shape)
    else: # Random create
      tensor = tf.random.uniform(shape, minval=min, maxval=max, dtype=tf.float32)
    return tensor
```

<h1>b. Compute the sum, mean, or standard deviation of all elements in a tensor.<h1>

```python
def compute_sum_mean_std(tensor, typeCal = 3): # 0 is only calculate sum, 1 is only calculate mean_val, 2 is only calculate std_val, 3 is calculate all
    total_sum = tf.reduce_sum(tensor)
    mean_val = tf.reduce_mean(tensor)
    std_val  = tf.math.reduce_std(tensor)

    if typeCal == 0: return total_sum
    elif typeCal == 1: return mean_val
    elif typeCal == 2: return std_val
    else: return total_sum, mean_val, std_val
```

<h1>c. Perform mathematical operations on tensors, such as finding the maximum or minimum value, or taking the dot product of two tensors.</h2>

```python
def maximum_in_tensor(tensor):
  return tf.reduce_max(tensor)

def minimum_in_tensor(tensor):
  return tf.reduce_min(tensor)

def dot_2_tensors(tensor1, tensor2):
  if tensor1.shape != tensor2.shape:
    print("\033[1;31mShapes must match!\033[0m")
    return None
  return tf.reduce_sum(tensor1 * tensor2)
```

<h1>d. Access and manipulate specific elements or slices of a tensor, and index and slice tensors based on certain conditions.</h1>


```python
def access_element(tensor, row_idx, col_idx):
  print(tensor[row_idx, col_idx].numpy())

def manipulate_element(tensor, row_idx, col_idx, newValue):
  t_var = tf.Variable(tensor)
  t_var[row_idx, col_idx].assign(newValue)
  t = tf.convert_to_tensor(t_var)
  print(f"Assigned newValue = {newValue} for element at [{row_idx},{col_idx}] : {t[row_idx, col_idx]}")
  return t

def access_by_slice(tensor, row_idx_start, row_idx_end, col_idx_start, col_idx_end):
  print(tensor[row_idx_start:row_idx_end, col_idx_start: col_idx_end])
```

<h1>e. Define and call functions in TensorFlow, and pass tensors as arguments to functions.<h1>

```python
@tf.function
def sum_2tensors(a, b):
  if a.shape != b.shape:
    print("\033[1;31mtensor A's shape must be equal to tensor B'shape!\033[0m")
    return None
  return a + b

@tf.function
def minus_2tensors(a, b):
  if a.shape != b.shape:
    print("\033[1;31mtensor A's shape must be equal to tensor B'shape!\033[0m")
    return None
  return a - b

@tf.function
def mul_2tensors(a, b):
  if a.shape != b.shape:
    print("\033[1;31mtensor A's shape must be equal to tensor B'shape!\033[0m")
    return None
  return a * b
```

```python
def supportInputTensors():
  while True:
    w = int(input("How many rows you want?(recommend = 1): "))
    h = int(input("How many cols you want?: "))
    inp = int(input("1 if input manually | 2 if random generate:"))
    if inp != 1 and inp != 2:
      print(f"\033[1;31moption should be 0 or 1 or 2!\033[0m")
      continue

    listValue = []
    if inp == 1: # user choose input manually
      while len(listValue) != w*h:
        value = int(input(f"Inputing value {len(listValue)} on {w*h} elements: "))
        listValue.append(float(value))

    shape = [w, h]
    tensor = nd_tensor_construct(listValue, shape)
    return tensor


def Run_Program():
  tensors = {}
  print("-------------- Adding n tensors is necessary for this program!")
  print("Recommending shape of tensor is [1,n] with 1 is w and n is h")
  while True:
    typeProblem = int(input("\n0.If you want to Exit\n1.If Adding Tensor\n2.If Compute the sum, mean, or standard deviation of all elements in a tensor. \n3.Perform mathematical operations on tensors, such as finding the maximum or minimum value, or taking the dot product of two tensors. \n4.Access and manipulate specific elements or slices of a tensor, and index and slice tensors based on certain conditions.\n5.Define and call functions in TensorFlow, and pass tensors as arguments to functions. \n6.If you want to print all tensors you input\nWhat requirement you want to run?: "))
    if typeProblem == 0:
      print("Thank for your using!")
      break
    elif typeProblem == 1:
      n_len = len(tensors)
      print(f"\nPlease give me the Tensor {n_len+1}(w*h):")
      in_tensor = supportInputTensors()
      tensors[n_len+1] = in_tensor
    elif typeProblem == 2:
      inputdict = {0: "total sum", 1: "mean val", 2: "std"}
      idx = int(input(f"\nInputing tensor you want to choose tensor[1 ..... {len(tensors)}]: "))
      if idx <0 or idx > len(tensors):
        print(f"\033[1;31moption should be  from 0 to {len(tensors)}!\033[0m")
        continue

      command = (int(input(f"Inputing 0 if calculating total_sum || 1 if calculating mean value || 2 if calculating standard deviation || 3 if calculating all:")))
      print(f"\nThe result after calculation for your tensor {idx}:")
      if command == 0 or command == 1 or command == 2:
        value = compute_sum_mean_std(tensors[idx], typeCal = command)
        print(f"{inputdict[command]}: {value.numpy():.3f}")
      elif command == 3:
        s, m, std = compute_sum_mean_std(tensors[idx], typeCal = command)
        print(f"total sum: {s.numpy():.3f}")
        print(f"mean val: {m.numpy():.3f}")
        print(f"standard deviation: {std.numpy():.3f}")
      else:
        print(f"\033[1;31moption should be 0 or 1 or 2 or 3!\033[0m")
        continue

    elif typeProblem == 3:
      command = (int(input(f"Inputing 0 if calculating maximum in tensor || 1 if calculating min in tensor || 2 if calculating dot 2 tensors:")))
      if command == 0 or command == 1:
        idx = int(input(f"\nInputing tensor you want to choose tensor[1 ..... {len(tensors)}]: "))
        if idx <0 or idx > len(tensors):
          print(f"\033[1;31moption should be from 1 to {len(tensors)}!\033[0m")
          continue
        if command == 0:
          max_value = maximum_in_tensor(tensors[idx])
          print(f"max value of tensor {idx}: {max_value:.3f}")
        elif command == 1:
          min_value = minimum_in_tensor(tensors[idx])
          print(f"min value of tensor {idx}: {min_value:.3f}")
      elif command == 2:
        print(f"\nTo calculate dot product, please providing me 2 tensors")
        idx1 = int(input(f"\nInputing first tensor[1 ..... {len(tensors)}]: "))
        idx2 = int(input(f"\nInputing second tensor[1 ..... {len(tensors)}]: "))
        dot_product_rs = dot_2_tensors(tensors[idx1], tensors[idx2])
        if dot_product_rs != None:
          print(f"dot product of tensor {idx1} and tensor {idx2}: {dot_product_rs:.3f}")
      else:
        print(f"\033[1;31moption should be 0 or 1 or 2 or 3!\033[0m")
        continue

    elif typeProblem == 4:
      command = (int(input(f"Inputing 0 if accessing tensor at specific index of row, col || 1 if manipulating tensor at specific index of row, col|| 2 if accessing by slices:")))
      idx = int(input(f"\nInputing tensor you want to choose tensor[1 ..... {len(tensors)}]: "))
      if command == 0 or command == 1:
        row_idx = int(input(f"\nInputing index of row you want to access(0 to {tensors[idx].shape[0]-1}):"))
        col_idx = int(input(f"Inputing index of col you want to access(0 to {tensors[idx].shape[1]-1}):"))
        if command == 0:
          access_element(tensors[idx], row_idx, col_idx)
        elif command == 1:
          newValue = float(input(f"Inputing newValue you want to change: "))
          tensors[idx] = manipulate_element(tensors[idx], row_idx, col_idx, newValue)
      elif command == 2:
        row_idx_start = int(input(f"\nInputing start index of row you want to access(0 to {tensors[idx].shape[0]}):"))
        row_idx_end = int(input(f"Inputing end index of row you want to access(0 to {tensors[idx].shape[0]}):"))
        col_idx_start = int(input(f"Inputing start index of col you want to access(0 to {tensors[idx].shape[1]}):"))
        col_idx_end = int(input(f"Inputing end index of col you want to access(0 to {tensors[idx].shape[1]}):"))
        access_by_slice(tensors[idx], row_idx_start, row_idx_end, col_idx_start, col_idx_end)

    elif typeProblem == 5:
      inputdict = {0: "sum_2tensors", 1: "minus_2tensors", 2: "mul_2tensors"}
      command = (int(input(f"Inputing 0 if using sum_2tensors() || 1 if using minus_2tensors() || 2 if using mul_2tensors:")))
      print(f"\nTo calculate sum_2tensors or minus_2tensors or mul_2tensors, please providing me 2 tensors")
      idx1 = int(input(f"\nInputing first tensor[1 ..... {len(tensors)}]: "))
      idx2 = int(input(f"Inputing second tensor[1 ..... {len(tensors)}]: "))
      if command == 0:
        rs = sum_2tensors(tensors[idx1], tensors[idx2])
      elif command == 1:
        rs = minus_2tensors(tensors[idx1], tensors[idx2])
      elif command == 2:
        rs = mul_2tensors(tensors[idx1], tensors[idx2])
      else:
        print(f"\033[1;31moption should be 0 or 1 or 2 or 3!\033[0m")
        continue

      if rs != None:
        print(f"result of {inputdict[command]} between tensor {idx1} and tensor {idx2} = \n{rs}")

    elif typeProblem == 6:
      print(f"\n Your input:")
      for k, value in tensors.items():
        print(f"Tensor {k}: {value}")
```
