<h1>Part2 Basic TensorFlow Question 10 points<h1>
<h1>a. Construct an n-dimensional tensor in TensorFlow.<h1>


```python
def nd_tensor_construct(listValue, shape, min = 0, max=100):
    if len(listValue) != 0: #For input from keyboard
      tensor = tf.constant(listValue)
      tensor = tf.reshape(tensor, shape)
    else: # Random create
      tensor = tf.random.uniform(shape, minval=min, maxval=max, dtype=tf.float32)
    return tensor
```

<h1>b. Compute the sum, mean, or standard deviation of all elements in a tensor.<h1>

```python
def compute_sum_mean_std(tensor, typeCal = 3): # 0 is only calculate sum, 1 is only calculate mean_val, 2 is only calculate std_val, 3 is calculate all
    total_sum = tf.reduce_sum(tensor)
    mean_val = tf.reduce_mean(tensor)
    std_val  = tf.math.reduce_std(tensor)

    if typeCal == 0: return total_sum
    elif typeCal == 1: return mean_val
    elif typeCal == 2: return std_val
    else: return total_sum, mean_val, std_val
```

<h1>c. Perform mathematical operations on tensors, such as finding the maximum or minimum value, or taking the dot product of two tensors.</h2>

```python
def maximum_in_tensor(tensor):
  return tf.reduce_max(tensor)

def minimum_in_tensor(tensor):
  return tf.reduce_min(tensor)

def dot_2_tensors(tensor1, tensor2):
  if tensor1.shape != tensor2.shape:
    raise ValueError("Shapes must match")
  return tf.reduce_sum(tensor1 * tensor2)
```

<h1>d. Access and manipulate specific elements or slices of a tensor, and index and slice tensors based on certain conditions.</h1>


```python
def access_element(tensor, row_idx, col_idx):
  print(tensor[row_idx, col_idx].numpy())

def manipulate_element(tensor, row_idx, col_idx, newValue):
  t_var = tf.Variable(tensor)
  t_var[row_idx, col_idx].assign(newValue)
  t = tf.convert_to_tensor(t_var)
  print(f"Assigned newValue = {newValue} for element at [{row_idx},{col_idx}] : {t[row_idx, col_idx]}")
  return t

def access_by_slice(tensor, row_idx_start, row_idx_end, col_idx_start, col_idx_end):
  print(tensor[row_idx_start:row_idx_end, col_idx_start: col_idx_end])
```

<h1>e. Define and call functions in TensorFlow, and pass tensors as arguments to functions.<h1>

```python
@tf.function
def sum_2tensors(a, b):
  if a.shape != b.shape:
    raise Exception("tensor A's shape must be equal to tensor B'shape")
  return a + b

@tf.function
def minus_2tensors(a, b):
  if a.shape != b.shape:
    raise Exception("tensor A's shape must be equal to tensor B'shape")
  return a - b

@tf.function
def mul_2tensors(a, b):
  if a.shape != b.shape:
    raise Exception("tensor A's shape must be equal to tensor B'shape")
  return a * b
```
